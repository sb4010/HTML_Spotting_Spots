<h2> Recommendation engine, plotting maps, finding nearest station </h2>

This code plots maps, spatially joins boroughs and census track, finds the nearest subway station, and provides recommendation based on three things - same rating, not a same census track and same category of restaurant.

<h4>Notes</h4>
1. This file takes only csv files and not xlsx files. Please make sure to convert xlsx files from previous code to csv files.
2. It is recommended to run this code with restaurant information from all boroughs otherwise the code will give recommendations of restaurants only within a borough.
3. Assumption is made that user will walk 0.3 miles from restaurant to station to make it the nearest station. However, this can be changed to 0.4 miles or any distance by changing it in the 'if' statement.
4. Only restaurants outside the same census tracks are recommended to diversify the crowd choices.However, this can be changed by commenting the 'if' statement.
5. Category matching is done even if only one category is matching between two restaurants.

```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import sin, cos, sqrt, atan2, radians
from pandas import ExcelWriter

venue_stuff = pd.read_csv("Combined_restaurant_categories.csv")

#Converting into GeoDataFrame and plotting restaurants against boroughs.
import geopandas
from shapely.geometry import Point
venue_stuff['Coordinates'] = list (zip (venue_stuff.venue_longitude, venue_stuff.venue_latitude))
venue_stuff["Coordinates"] = venue_stuff["Coordinates"].apply(Point)

gdf = geopandas.GeoDataFrame(venue_stuff, geometry='Coordinates')

boros=geopandas.read_file("boros.geojson")

gdf.plot(figsize=(10, 50),color='red')
base = boros.plot(figsize=(50, 50), color='lightgrey', edgecolor='white')
gdf.plot(ax=base, marker='o', color='red', markersize=10);


#Doing spatial join of venues with census tracks
#Plotting restaurants against census tracks

census_tracks=geopandas.read_file("Censustracks.geojson")
gdf.plot(figsize=(10, 50),color='red')
base2 = census_tracks.plot(figsize=(50, 50), color='lightgrey', edgecolor='white')
gdf.plot(ax=base2, marker='o', color='red', markersize=10);


gdf.crs = {"init" :"epsg:4326"}
venues_with_ct = geopandas.sjoin (gdf, census_tracks, how = "inner", op = "intersects")
print (venues_with_ct)


#Doing spatial join of stations with census tracks
stations = pd.read_csv("Subway_station.csv")
stations["lon"] = stations['the_geom'].str[7:23]
stations["lat"] = stations['the_geom'].str[25:-1]

stations["lat"]=pd.to_numeric(stations["lat"],errors='raise')
stations["lon"]=pd.to_numeric(stations["lon"],errors='raise')

clat = np.array([])

#correction in string parsing 
for index, row in stations.iterrows():
    temp_val = 0
    if row["lat"] < 40.00:
        temp_val = row["lat"] + 40
    else:
        temp_val = row["lat"]
    clat = np.append (clat, temp_val)

stations["clat"] = clat



stations['Coordinates'] = list(zip(stations.lon, stations.clat))
stations['Coordinates'] = stations['Coordinates'].apply(Point)


gdf_stations = geopandas.GeoDataFrame(stations, geometry='Coordinates')
#print (gdf_stations)
#plotting the map

gdf_stations.plot(figsize=(10, 50),color='red')
gdf_stations.plot(ax=base2, marker='o', color='red', markersize=10);

gdf_stations.crs = {"init" :"epsg:4326"}
stations_with_ct = geopandas.sjoin (gdf_stations, census_tracks, how = "inner", op = "intersects")
print (stations_with_ct)


#finding nearest subway station

#compute distance between every restaurant and subway station and find nearest subway stations
count = 0
R = 6373.0   #Radius of Earth
ct = 0.01745329   #Degree to radian conversion
station_list2 = np.array([])

#Distance calculation
for index2, row2 in venues_with_ct.iterrows():
    found_station = False
    nearest_stations = ' '
    for index1, row1 in stations_with_ct.iterrows():
        ff = float (row1['clat'])
        lat1 = ff * ct
        b = float (row1['lon'])
        lon1 = b * ct
        lat2 = radians(row2['venue_latitude'])
        lon2 = radians(row2['venue_longitude']) 
        dlon = lon2 - lon1
        dlat = lat2 - lat1     
        a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
        c = 2 * atan2(sqrt(a), sqrt(1 - a))
        distance = R * c
        
        #Assumption every person wont walk more than 0.3 miles(0.4828032km) from restaurant to subway station. 
        #if distance < 0.643738:
        if distance < 0.4828032:
                       
            found_station = True
            nearest_stations = nearest_stations + row1["NAME"] + ','
             
    station_list2 = np.append (station_list2, nearest_stations)

    
        
print (station_list2)
print (station_list2.size)


#Adding nearest stations to DataFrame

venues_with_ct ['nearest_stations'] = station_list2
venue_recommendations = np.array([])

category_array = np.array([])

#Recommendation Engine
for index, row in venues_with_ct.iterrows():
    A_ct = row['ct2010']
    A_rating = row['rating']
    catgory = row['categories']
    category_list1 = catgory.split (",")
    #Removing last comma ','
    category_list1 = category_list1[:-1]
    recommendations = ' '
    
    
    for index1, row1 in venues_with_ct.iterrows():
        #Checking if not in same census track
        if row1['ct2010'] != A_ct:
            #Checking if same rating
            if row1['rating'] == A_rating:
                category2 = row1['categories']
                category_list2 = category2.split(',')
                category_list2 = category_list2[:-1]
                #Checking if any category matches
                for category_1 in category_list1:
                    for category_2 in category_list2:
                        #print (category_1)
                        #print (category_2)
                        if category_1 == category_2:
                            
                            recommendations = recommendations + row1['venue_name'] + ','
    venue_recommendations = np.append (venue_recommendations, recommendations)


venues_with_ct ['Recommmendations'] = venue_recommendations
    
    

print(venues_with_ct)

venues_final = pd.DataFrame()
#Taking only useful and relevant fields
venues_final = venues_with_ct.filter (items = ['venue_id', 'venue_name', 
                                               'venue_location', 'venue_latitude', 'venue_longitude', 
                                               'rating', 'categories', 'ntacode', 'ntaname',
                                              'boro_name', 'ct2010', 'boro_code',
                                              'nearest_stations', 'Recommmendations'])

#Writing to excel sheet
writer = ExcelWriter('Combined_restaurant_recco.xlsx')
venues_final.to_excel(writer, 'Sheet 1')
writer.save()



```
