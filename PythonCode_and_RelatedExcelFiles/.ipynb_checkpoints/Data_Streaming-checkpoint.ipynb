<h2> Streaming of Foursquare Data </h2>

This code streams data from Foursquare API of the hundred best restaurant in Manhattan. The same code was run on Queens, Brooklyn and Bronx as well to get the data of 100 best restaurants in the respective boroughs.

<h4> Why Foursquare API? </h4>

We have used foursquare instead of other sources because:

1. It provides 99,500 Regular Calls / Day, 500 Premium Calls / Day, 1 Photo per Venue, 1 Tip per Venue.
2. It has a recomendation engine that provides relevant data of places within the boundary mentioned.
3. Provides limit parameter(which can go till 200) that lets you search for large list of places.
4. Allows nested calls.

<h4> Note </h4>

1. We streamed data of only first 100 restaurants for exclusivity.
2. The output of this file is a JSON file.

```python
import json, requests
url = 'https://api.foursquare.com/v2/venues/explore'

params = dict(
  client_id='0MJHD4DIWITUATX3E5ZVUCGSJ42UMG1XHWLI1MJNU1XQDWI3',
  client_secret='CRYCX2A5OQUVQZO4LJ55AYOY3K3TUZXJU1IQLR521TKXATFO',
  v='20180323',
  near="Manhattan",
  query='Restaurant',
  limit=50
)
resp = requests.get(url=url, params=params)
data = json.loads(resp.text)
print(data)

with open('restaurant_Manhattan.json', 'a') as tf:
            # Write a new line
            tf.write('\n')

            # Write the json data directly to the file
            json.dump(data, tf)
```
