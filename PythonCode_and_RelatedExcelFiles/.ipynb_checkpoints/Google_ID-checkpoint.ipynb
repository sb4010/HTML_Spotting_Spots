<h2> Finding Google ID of restaurants </h2>

This code finds google IDs of restaurants and adds them into excel sheets.

<h4>Why do we need google ID?</h4>
Google API gives restaurant information but doesn't give it in bulk unlike foursquare API. Although it is possible to get wait time and crowd estimation of a restaurant using google map. There are following challenges :
    1)Google map requires google ID to get any information.
    2)Google map API is available in javascript so we need googleplaces python wraper around it which requires usage of   googleplaces library.
        
Installation Instructions and documentation can be found out at https://github.com/slimkrazy/python-google-places
In this code we have used name of the restaurant as a keyword to find google ID of the place.This is our keyword search.
    
<h4>Notes</h4>
1. This file takes only csv files and not xlsx files. Please make sure to convert xlsx files from previous code to csv files.
2. It is not possible to find google ID of every restaurant using keyword search so some restaurant IDs were found out manually.Usually the success rate of keyword search is 90-98%. Another way of doing this is using address of the restaurant as the keyword instead name in code described in last cell.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import sin, cos, sqrt, atan2, radians
from pandas import ExcelWriter


from googleplaces import GooglePlaces, types, lang
#Googleplaces API key
API_KEY = 'AIzaSyAcprvCxEEwlfaxBSKpCrUPb9tGAFElyBg'
places_name = np.array([])

venue_stuff = pd.read_csv("Combined_restaurant_reco.csv")
test = pd.DataFrame()
test['name'] = venue_stuff['venue_name']
test['latitude'] = venue_stuff['venue_latitude']
test['longitude'] = venue_stuff['venue_longitude']

google_places = GooglePlaces(API_KEY)
googleIDs = np.array([])
googleNames = np.array([])
place_count = 0

#Calling googleplaces API for every restaurant
for index, row in test.iterrows():
   #Nearby search is used here to find restaurant with given name incase if there are multiple restaurants with same names query result will contain more than one elements. 
    query_result = google_places.nearby_search(
    location='New York', keyword=row['name'], radius = 100000)
    
    #print(query_result)
    place_found = False
    #Enter here for query result with multiple values. 
    for place in query_result.places:
        if place.name:
            
            location = place.geo_location
            lat = location['lat']
            long = location['lng']
            #Making sure that the restaurant is the restaurant that we want. Hence, use Lat, Long based filtering.
            #Values are caliberated after multiple testings.
            if -0.001 < row['latitude']-float(lat) < 0.001 and -0.001 < row['longitude']-float(long) < 0.0031:
                
                googleIDs = np.append (googleIDs, place.place_id)
                googleNames = np.append (googleNames, place.name)
                place_found = True
                place_count = place_count+1
                break
                
    if place_found == False:
        googleIDs = np.append (googleIDs, " ")
        googleNames = np.append (googleNames, " ")
                
venue_stuff ['PlaceID'] = googleIDs

print (venue_stuff)
#Writing to excel sheet
writer = ExcelWriter('Combined_restaurant_google.xlsx')
venue_stuff.to_excel(writer, 'Sheet 1')
writer.save()



```
